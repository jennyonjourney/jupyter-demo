# Downloading and Installing Anaconda and Jupyter Notebook

1. Download 64-bit Anaconda from https://www.anaconda.com/products/individual
    - Anaconda is a package manager tool that allows sets of libraries to be confined to environments. This helps to prevent conflict between installed packages for different projects
    - Make Anaconda your default python tool per the install prompt
    - Do not add Anaconda to PATH


2. Once installed, open terminal
    - For Mac and Linux this is the standard terminal that is already integrated with your system
    - For Windows, you'll need to use the Anaconda Prompt Terminal. 
    
    
3. Create a new environment for the course
    - conda create --name environmentName python=3.7 
   
   
4. Activate your new environment
    - conda activate environmentName
  
  
5. Install jupyter notebook
    - conda install jupyter 
    
    
6. Install other useful packages relevant to the class
    - conda install pandas (dataframe tool) https://pandas.pydata.org/docs/user_guide/10min.html#min
    - conda install matplotlib (plotting) https://matplotlib.org/3.1.1/tutorials/index.html
    - conda install scikit-learn (out-of-the-box ML tools) https://scikit-learn.org/stable/index.html
    - numpy and scipy are installed along side scikit-learn. These libraries include useful linear algebra and statistics tools.


7. Open Jupyter notebooks
    - navigate in terminal to your desired working directory
    - simply type "jupyter notebook" into the console


8. Guide on how to use Jupyter notebook
    - https://www.dataquest.io/blog/jupyter-notebook-tutorial/
    


## Importing Libraries

Libraries are publicly available repositories of code. They can be imported into your projects by using the *import* keyword. You can also give a library an alias by using the *as* keyword.

```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LinReg
```

```python
X = np.linspace(0,10,1000)
eps = rand.normal(loc=0,scale=10,size=1000)
y = (1.245*X**2 + eps)

X = X.reshape(-1,1)
y = y.reshape(-1,1)

model=LinReg()
model.fit(X,y)
y_pred = model.predict(X)

plt.scatter(X,y,label='synthetic data')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(X,y_pred,color='red',label='prediction')
plt.title('Python Linear Regression Demo, coef: %.2f, intercept: %.2f'
          %(model.coef_,model.intercept_))
plt.legend()
plt.show()
```
